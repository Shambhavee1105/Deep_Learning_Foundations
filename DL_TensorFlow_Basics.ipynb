```python
import tensorflow as tf
#1.examples using TensorFlow
#scalar tensor
t1 = tf.constant(4)
print("Scalar Tensor: ")
print(t1)
```

```python
#vector tensor
t2 = tf.constant([1,5,10,15,20,25])
print("Vector Tensor: ")
print(t2)
```

```python
#matrix tensor
t3 = tf.constant([[1,5], [10,15], [20,25]])
print("Matrix Tensor: ")
print(t3)
```

```python
#3D tensor
t4 = tf.constant([[[1,5], [10,15]], [[20,25], [30,35]]])
print("3D Tensor: ")
print(t4)
```

```python
#zeros tensor
t5 = tf.zeros([3,3])
print("Zeros Tensor: ")
print(t5)
```

```python
#ones tensor
t6 = tf.ones([2,4])
print("Ones Tensor: ")
print(t6)
```

```python
#matrix tensor
t7 = tf.constant([[2,4], [6,8], [10,12]])
print("Matrix Tensor: ")
print(t7)
```

```python
#2.addition of 2D tensors
t_add = tf.add(t3, t7)
print("Added matrices t3 and t7: ")
print(t_add)
```

```python
#3.matrix operations using tensors
t8 = tf.constant([[2,4], [6,8], [10,12]])
t9 = tf.constant([[3,5], [7,9], [11,13]])
print("Matrix Tensors: ")
print(t8)
print(t9)

#a.addition
ta = tf.add(t8, t9)
print("Added matrices: ")
print(ta)

#b.subtraction
tb = tf.subtract(t8, t9)
print("Subtracted matrices: ")
print(tb)

#c.element wise multiplication
tc = tf.multiply(t8, t9)
print("Element-wise multiplied matrices: ")
print(tc)

#d.multiplication(dot product)
td = tf.matmul(t8, tf.transpose(t9))
print("Dot product of matrices: ")
print(td)

#e.transpose
te = tf.transpose(t8)
print("Transpose of matrix t8: ")
print(te)
```
